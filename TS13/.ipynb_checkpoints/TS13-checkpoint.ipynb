## Simulaciones
### Ejercicio 1

```python
import sympy as sp
import splane as tc2
from schemdraw import Drawing
from schemdraw.elements import  Resistor, Capacitor, Inductor
import numpy as np

s = sp.symbols('s ', complex=True)

ZZ = (s+2)*(s+6)*6/(s+1)/(s+4)
a = (43 + np.sqrt(489))/10
Z2, ZR2 = tc2.remover_valor(ZZ, sigma_zero = a)

Z4, Z3, R3, C3 = tc2.remover_polo_sigma(Z2, sigma = 1, isImpedance = True, isRC = True)

Z6, Z5, R5, C5 = tc2.remover_polo_sigma(Z4, sigma = 4, isImpedance = True, isRC = True)

Y8, YR7 = tc2.remover_valor_en_infinito(1/Z6)

d = Drawing(unit=4)

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Z',  
                                          YR7, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)


d = tc2.dibujar_elemento_derivacion(d, Resistor, YR7)

d = tc2.dibujar_tanque_RC_serie(d, R5, C5)

d = tc2.dibujar_tanque_RC_serie(d, R3, C3)

d = tc2.dibujar_elemento_serie(d, Resistor, ZR2)

display(d)
```

### Ejercicio 2

```python
import sympy as sp
import splane as tc2
from schemdraw import Drawing
from schemdraw.elements import  Capacitor, Inductor, Resistor

# Resolucin simblica

s = sp.symbols('s ', complex=True)

ZZ = (2*s**2+1)/(s**3+2*s)

Y2, Yc3 = tc2.remover_polo_infinito(1/ZZ, omega_zero = 3 )

C3= Yc3/s

Z4, Zt2, L2, C2 = tc2.remover_polo_jw(1/Y2, omega = 3 , isImpedance = True)

Y6, YC1 = tc2.remover_polo_infinito(1/Z4)

C1 = YC1/s
d = Drawing(unit=4)

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Z',  
                                          ZZ, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C1)

d = tc2.dibujar_tanque_serie(d, L2, C2)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C3)


display(d)
```
