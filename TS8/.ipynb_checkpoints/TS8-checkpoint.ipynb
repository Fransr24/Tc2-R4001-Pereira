3)a) Para la media aritmetica el promedio de todos los coeficientes tienen que dar 1, por lo que cada componente deberia representar bx/cant de muestras, siendo bx los distintos coeficientes de la salida.
La frecuencia que se deberia adoptar es una de 200hz, asi nyquist vale 100 que es el doble de 50, con eso seria fuficiente para filtrar a esa frecuencia. \
b) Los sistemas demoran pi/2 y para que se comporte como un derivador ideal con una tolerancia del 5% tengo que tener una frecuencia de nyquist que lo cumpla, para conseguirlo debo comparar con el modelo analogico y ver como se va comportando en las frecuencias de interes, ir disminuyendo la frecuencia de sampleo hasta obtener el 5%


Los graficos los muestro en la simulacion en phyton usando una funcion que se encargar√° de hacer las cosas necesarias para ver los graficos e ir repitiendolo en cada punto

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt

def Graficos_TS8(fc, fs, xlim1=0, xlim2=10, ylim1=-60, ylim2=4):
    WN = int(5*10e3) # Cantidad de puntos para la funcion Freqz (por el procesamiento digital)
    KHz = 1000
    
    #defino nyquist y la frecuencia de sampling (fs argumento de la funcion)
    nq = fs/2
    ws = 2*np.pi*fs
    #defino norma de frecuencia y normalizo
    wo = 2*np.pi*fc
    fs_n = fs/wo
    ws_n = ws/wo
    #asumo sin prewarping
    K = 2*fs_n
    #Del butter orden 2
    Q = np.sqrt(2)/2
    
    # Coeficientes desnormalizados del numerador y denominador
    #Est√° calculado en la hoja pero de esta manera va a ser mas precisa
    A0 = K**2 + K/Q + 1
    A1 = 2 - 2*K**2
    A2 = K**2 - K/Q + 1
    B0 = 1
    B1 = 2
    B2 = 1
    #Coeficientes normalizados
    a0 = A0/A0
    a1 = A1/A0
    a2 = A2/A0
    b0 = B0/A0
    b1 = B1/A0
    b2 = B2/A0
    #Creo la funcion transferencia con los coeficientes
    numz = [b0, b1, b2]
    denz = [a0, a1, a2]
    w, h = sig.freqz(numz,denz,WN)
    
    ##Ahora me encargo de definir el pasa bajos analogico y pasarlo con la transformada bilineal
    num_lp = [0,  0,  1]
    den_lp = [1, 1/Q, 1]
    #Digital hecho por phyton
    numz_py, denz_py = sig.bilinear(num_lp, den_lp, fs_n) 
    w_py, h_py = sig.freqz(numz_py,denz_py,WN)
    #De manera analogica por phyton
    w_analog, h_analog = sig.freqs(num_lp,den_lp,  worN=np.logspace(-3, 2, 1000))
    #Lo expreso en db
    h_analog_db    = 20*np.log10(np.maximum(np.abs(h_analog), 1e-4))
    h_db    = 20*np.log10(np.maximum(np.abs(h), 1e-4))
    h_db_py = 20*np.log10(np.maximum(np.abs(h_py), 1e-4))
    
    
    #Funciones para graficar
    fig, ax = plt.subplots(nrows=1, ncols=1, sharex=True)
    fig.set_size_inches(16,12)
    # MODULO
    fig.suptitle('ButterWorth N=2')
    #muestro las 3 distintas formas que lo calcul√© 
    ax.plot(w*nq/(np.pi*KHz), h_db,linewidth=3, label = 'Calculado') #Digital que calcul√© yo
    ax.plot(w_py*nq/(np.pi*KHz), h_db_py,color='purple',linestyle='dashed',linewidth=3, label = 'Python') #verificacion de digit
    ax.plot(w_analog*fc/KHz, h_analog_db, color = 'orange',linestyle=':',  linewidth=3, label = 'Analogico')#Analogico phyton
    ax.grid(True)
    ax.set_title('Respuesta Magnitud $f_c = {} KHz$ | $f_s = {}KHz$'.format(fc/KHz, fs/KHz))
    ax.set_ylabel('Magnitud [dB]')
    ax.set_xlabel('Frecuencia [KHz]')
    ax.legend()
    ax.set_xlim([xlim1, xlim2])
    ax.set_ylim([ylim1, ylim2])
    
#Primer caso
fc = 1e3
fs = 100e3
Graficos_TS8(fc, fs,0,2,-6,1)
```

Como se explica en el codigo, esta simulacion calcula la transferencia de 3 maneras, a mano que es lo mas parecido a lo que calcul√© en papel, hecho completamente por phyton para verificar con las cuentas que est√°n a mano y el filtro analogico(Tambien phyton). Como se puede ver los 3 son iguales, ya que Fs que es la frecuencia de sampleo vale 100K, por lo que nyquist valen 50K, con una frecuencia de corte de 1K no estamos ni cerca de nyquist

### B)

```python
fc = 1e3
fs = 10e3
Graficos_TS8(fc, fs,0,2,-6,1)
```

Ac√° se aprecia que como nos encontramos a frecuencias mas cercanas a Nyquist, poco a poco la respuesta del filtro digital se va separando de la analogica y no se cumplen los 3db en la frecuencia de corte

### C Pasa altos caso A

```python
def Graficos_TS8_Hp(fc, fs, xlim1=0, xlim2=10, ylim1=-60, ylim2=4):
    #Muy parecido a lo otro solo que pasa altos
    WN = int(5*10e3)
    KHz = 1000
    
    nq = fs/2
    ws = 2*np.pi*fs
    wo = 2*np.pi*fc
    fs_n = fs/wo
    ws_n = ws/wo
    K = 2*fs_n
    Q = np.sqrt(2)/2
    
    num_hp = [1,  0,  0]
    den_hp = [1, 1/Q, 1]
    #Digital hecho por phyton
    numz_py, denz_py = sig.bilinear(num_hp, den_hp, fs_n) 
    w_py, h_py = sig.freqz(numz_py,denz_py,WN)
    #De manera analogica por phyton
    w_analog, h_analog = sig.freqs(num_hp,den_hp,  worN=np.logspace(-3, 2, 1000))
    #Lo expreso en db
    h_analog_db    = 20*np.log10(np.maximum(np.abs(h_analog), 1e-4))
    h_db_py = 20*np.log10(np.maximum(np.abs(h_py), 1e-4))
    
    
    #Funciones para graficar
    fig, ax = plt.subplots(nrows=1, ncols=1, sharex=True)
    fig.set_size_inches(16,12)
    # MODULO
    fig.suptitle('ButterWorth N=2')
    #muestro las 3 distintas formas que lo calcul√© 
    ax.plot(w_py*nq/(np.pi*KHz), h_db_py,color='purple',linestyle='dashed',linewidth=3, label = 'Python') #verificacion de digit
    ax.plot(w_analog*fc/KHz, h_analog_db, color = 'orange',linestyle=':',  linewidth=3, label = 'Analogico')#Analogico phyton
    ax.grid(True)
    ax.set_title('Respuesta Magnitud $f_c = {} KHz$ | $f_s = {}KHz$'.format(fc/KHz, fs/KHz))
    ax.set_ylabel('Magnitud [dB]')
    ax.set_xlabel('Frecuencia [KHz]')
    ax.legend()
    ax.set_xlim([xlim1, xlim2])
    ax.set_ylim([ylim1, ylim2])
    
fc = 6e3
fs = 100e3
Graficos_TS8_Hp(fc, fs,0,12,-6,1)
```

Aca podemos ver que aunque haya aumentado la frecuencia de corte unas 6 veces, la diferencia entre analogico y digital sigue siendo muy peque√±a, pero ahora notoria, ya que la Frecuencia de corte aument√≥ 6 veces

```python
fc = 6e3
fs = 10e3
Graficos_TS8_Hp(fc, fs,0,12,-6,1)
```
