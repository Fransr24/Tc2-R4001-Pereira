### b)
Como se explic en el punto a el parametro S21 representa la transferencia directa de la red y en el caso del parametro S11, este representa el coeficiente de reflexion de la onda en el puerto 1, y dados los puntos de interes de la frecuencia podemos llegar a que:
- En w=0, S21=1 que es el centro de la banda de paso. Mientras que S11=0 significa que la onda no se refleja, es decir que est completamente adaptada.
- En w=1, S21 est en la frecuencia de corte. Mientras que S11=1, que quiere decir que la onda est desadaptada
- En w=inf, S21 = 0 cumpliendo su comportamiento de filtro pasa bajos. Mientras que S11=1 que quiere decir que la onda est desadaptada y se refleja completamente.
