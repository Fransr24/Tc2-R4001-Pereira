### Cómo podría obtener un circuito pasabanda con los mismos componentes originales y con qué parámetros quedaría diseñado:
La manera con la cual podrias obtener un circuito pasa Banda es Tomando como salida la tension que sale por el primer amplificador operacional. Sabiendo que en el camino entre la tension Va (1er operacional) y Vo (segundo operacional, salida que tomamos inicialmente) hay un proceso de integracion gracias al capacitor, llegamos a la conclusion de que si no recorremos este camino, no ocurriria dicho proceso, es decir que tendremos una S en el numerador de la transferencia. Con esta S en el numerador, el comportamiento de la transferencia pasa a ser el de un filtro Pasa banda.
Para ser mas exactos, no solo hay un proceso de integración, la relacion que tienen Va y Vo es Vo=
$\frac{Va}{SCR3}$

Por lo que al volver a calcular la transferencia el valor de K ya no seria $\frac{R3}{R1}$, sino que pasaria a ser uno que dependa de C y R1
