```python
import sympy as sp
import splane as tc2
from schemdraw import Drawing
from schemdraw.elements import  Resistor, Capacitor, Inductor
import numpy as np

# Resolución simbólica

s = sp.symbols('s ', complex=True)

# Sea la siguiente función de excitación
Imm = (s**5 + 12*s**3 + 20*s)/(s**6 + 26*s**4+125*s**2+100)

# Implementaremos Imm mediante Foster
k0, koo, ki = tc2.foster(Imm)

# Tratamos a nuestra función imitancia como una Y
tc2.dibujar_foster_derivacion(k0, koo, ki, y_exc = Imm)
```

```python
# Foster derivacion 

# Resolución simbólica

s = sp.symbols('s ', complex=True)

YY = s*(s**2+2)*(s**2+4)/(s**2+1)/(s**2+3)

Y2, Yc1 = tc2.remover_polo_infinito(YY)

# Yc1 es la admitancia removida
# extraigo C1
C1 = Yc1/s

Y4, Yt2, L2, C2 = tc2.remover_polo_jw(Y2, isImpedance = False)

Y6, Yt3, L3, C3 = tc2.remover_polo_jw(Y4, isImpedance = False) 

# Dibujamos la red resultante:
    
d = Drawing(unit=4)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Y',  
                                          YY, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C1)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_tanque_derivacion(d, L2, C2)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_tanque_derivacion(d, L3, C3)

display(d)
```

```python
# Foster serie
ZZ = (s**2+1)*(s**2+3)/s/(s**2+2)/(s**2+4) 

Z2, Zc1 = tc2.remover_polo_dc(ZZ)

# Zc1 es la impedancia removida
# extraigo C1
C1 = 1/(s*Zc1)

Z4, Zt2, L2, C2 = tc2.remover_polo_jw(Z2, isImpedance = False) 

Z6, Zt3, L3, C3 = tc2.remover_polo_jw(Z4, isImpedance = False) 

# Dibujamos la red resultante:
    
d = Drawing(unit=4)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Z',  
                                          ZZ, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_elemento_serie(d, Capacitor, C1)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_tanque_serie(d, L2, C2)

d = tc2.dibujar_espacio_derivacion(d)

d = tc2.dibujar_tanque_serie(d, L3, C3)

display(d)
```

```python
# Cauer 1

YY = s*(s**2+2)*(s**2+4)/(s**2+1)/(s**2+3)

Y2, Yc1 = tc2.remover_polo_infinito(YY)

# Yc1 es la admitancia removida
# extraigo C1
C1 = Yc1/s

Z4, Zl1 = tc2.remover_polo_infinito(1/Y2)

# Zl1 es la impedancia removida
# extraigo L1
L1 = Zl1/s

Y6, Yc2 = tc2.remover_polo_infinito(1/Z4) 

# Yc2 es la admitancia removida
# extraigo C2
C2 = Yc2/s

Z8, Zl2 = tc2.remover_polo_infinito(1/Y6)

# Zl2 es la impedancia removida
# extraigo L2
L2 = Zl2/s

Y10, Yc3 = tc2.remover_polo_infinito(1/Z8) 

# Yc3 es la admitancia removida
# extraigo C3
C3 = Yc3/s

# Dibujamos la red resultante:
    
d = Drawing(unit=4)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Y',  
                                          YY, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C1)

d = tc2.dibujar_elemento_serie(d, Inductor, L1)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C2)

d = tc2.dibujar_elemento_serie(d, Inductor, L2)

d = tc2.dibujar_elemento_derivacion(d, Capacitor, C3) 

display(d)
```

```python
# Cauer 2

ZZ = (s**2+1)*(s**2+3)/s/(s**2+2)/(s**2+4) 

Z2, Zc1 = tc2.remover_polo_dc(ZZ)

# Zc1 es la admitancia removida
# extraigo C1
C1 = 1/(s*Zc1)

Y4, Yl1 = tc2.remover_polo_dc(1/Z2)

# Yl1 es la impedancia removida
# extraigo L1
L1 = 1/(s*Yl1)

Z6, Zc2 = tc2.remover_polo_dc(1/Y4) 

# Zc2 es la admitancia removida
# extraigo C2
C2 = 1/(s*Zc2)

Y8, Yl2 = tc2.remover_polo_dc(1/Z6)

# Yl2 es la impedancia removida
# extraigo L2
L2 = 1/(s*Yl2)

Z10, Zc3 = tc2.remover_polo_dc(1/Y8) 

# Zc3 es la admitancia removida
# extraigo C3
C3 = 1/(s*Zc3) 

# Dibujamos la red resultante:
    
d = Drawing(unit=4)  # unit=2 makes elements have shorter than normal leads

d = tc2.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V$', '-'), 
                        current_lbl = '$I$')

d, zz_lbl = tc2.dibujar_funcion_exc_abajo(d, 
                                          'Z',  
                                          ZZ, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)

d = tc2.dibujar_elemento_serie(d, Capacitor, C1)

d = tc2.dibujar_elemento_derivacion(d, Inductor, L1)

d = tc2.dibujar_elemento_serie(d, Capacitor, C2)

d = tc2.dibujar_elemento_derivacion(d, Inductor, L2)

d = tc2.dibujar_elemento_serie(d, Capacitor, C3) 

display(d)
```
